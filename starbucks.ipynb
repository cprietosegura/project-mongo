```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()
import pandas as pd
import json
```

```python
yelp_api_key = os.getenv("YELP_API_KEY")
#coords= "37.4598046,-122.142886"
if not yelp_api_key:
    raise ValueError("NECESITAS UN TOKEN")
else:
    print("We have a yelp api key: ", yelp_api_key[0:4])

url='https://api.yelp.com/v3/businesses/search'
headers = {'Authorization': 'Bearer %s' % yelp_api_key}
 
#Ã¬n the dictionary, term can take values like food, cafes or businesses like McDonalds

params = {'term':'Starbucks','location':'Palo Alto'}
req=requests.get(url, params=params, headers=headers)
print('The status code is {}'.format(req.status_code))
```

```python
# printing the text from the response 
starbucks=json.loads(req.text)
```
